**Problem 1**

Part A

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def forward_euler_SIS(gamma, beta, delta_t, I_0, S_0):
    S = [S_0]
    I = [I_0]
    
    for t in range(1,25):
        s_t = S[t-1] + delta_t*((-beta*S[t-1]*I[t-1])+(gamma*I[t-1]))
        S.append(s_t)
        
        i_t = I[t-1] + delta_t*((beta*S[t-1]*I[t-1])-(gamma*I[t-1]))
        I.append(i_t)
    
    return S, I
```

```python
gamma = 2
beta = 3
S_0 = 0.99
I_0 = 0.01
R_0 = beta/gamma 
delta_ts = [2,1,0.5]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))

for i, delta_t in enumerate(delta_ts):
    S, I = forward_euler_SIS(gamma, beta, delta_t, I_0, S_0)
    t = np.arange(0,25, delta_t)
    analytical = (1-(1/R_0))/(1+((1-(1/R_0)-I_0)/(I_0))*np.exp(-(beta-gamma)*t))
    axes[i].plot(I, label='Forward Euler', color='red')
    axes[i].plot(t, analytical, label='Analytical', color='black', linestyle='dashed')
    axes[i].set_xlabel("time")
    axes[i].set_ylabel("people")
    axes[i].set_title(f'delta_t = {delta_t}')
    axes[i].set_xlim([0, 25])
    axes[i].set_ylim([0, 0.5])
    axes[i].legend()

plt.tight_layout()
plt.savefig("Figures/SIS Figures")
plt.show()
```

Part C

```python
def max_absolute_error(beta, gamma, S_0, I_0, delta_t):
    S, I = forward_euler_SIS(gamma, beta, delta_t, I_0, S_0)
    T = len(I)
    t = np.linspace(0, T*delta_t, T)
    analytical = (1-(1/R_0))/(1+((1-(1/R_0)-I_0)/(I_0))*np.exp(-(beta-gamma)*t))
    return np.max(np.abs(np.array(I) - analytical))

# Example
delta_t = 0.5
max_error = max_absolute_error(beta, gamma, S_0, I_0, delta_t)
print(f"Expected maximum absolute error for dt={delta_t}: {max_error}")
```

Part D

```python
delta_ts = [2,1,1/2,1/4,1/8,1/16,1/32]

max_errors = [max_absolute_error(beta, gamma, S_0, I_0, delta_t) for delta_t in delta_ts]

plt.figure(figsize=(8, 6))
plt.loglog(delta_ts, max_errors, marker='o')
plt.xlabel('Time Step Size (delta_t)')
plt.ylabel('Maximum Absolute Error')
plt.title('Maximum Absolute Error vs Time Step Size')
plt.grid(True)
plt.savefig('Figures/MaxAbsoluteError_byTimeStep')
plt.show()
```

```python
def forward_euler_ANM(gamma, beta, delta_t, I_0, S_0, time, V_0):
    S = [S_0]
    I = [I_0]
    R = [0]
    V_nothing = [V_0]
    
    for t in range(1,time):
        s_t = S[t-1] + delta_t*(-beta*S[t-1]*I[t-1])
        S.append(s_t)
        
        i_t = I[t-1] + delta_t*((beta*S[t-1]*I[t-1])-(gamma*I[t-1]) + (beta*V_nothing[t-1]*I[t-1]))
        I.append(i_t)
        
        r_t = R[t-1] + delta_t*(gamma*I[t-1])
        R.append(r_t)
        
        v_t = V_nothing[t-1] + delta_t*(-beta*V_nothing[t-1]*I[t-1])
        V_nothing.append(v_t)
        
    return S,I,R,V_nothing

def forward_euler_LM(gamma, beta, delta_t, I_0, S_0, time, V_0, VE):
    S = [S_0]
    I = [I_0]
    R = [0]
    V = [V_0]
    
    for t in range(1,time):
        s_t = S[t-1] + delta_t*(-beta*S[t-1]*I[t-1])
        S.append(s_t)
        
        i_t = I[t-1] + delta_t*((beta*S[t-1]*I[t-1])-(gamma*I[t-1]) + (beta*V[t-1]*I[t-1]*(1-VE)))
        I.append(i_t)
        
        r_t = R[t-1] + delta_t*(gamma*I[t-1])
        R.append(r_t)
        
        v_t = V[t-1] + delta_t*(-beta*V[t-1]*I[t-1]*(1-VE))
        V.append(v_t)
        
    return S,I,R,V
```

```python
beta_gamma = [(3,1), (4,1), (5,1)] #beta/gamma needs to equal R_0
S_0 = 0.99
I_0 = 0.01
R_0 = [3,4,5]
time = 100
v = 0.5
VE = 0.8
V_0 = v*(1-VE)
delta_t = 0.1

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
#ALM
for i, (beta, gamma) in enumerate(beta_gamma):
    S, I, R, V = forward_euler_ANM(gamma, beta, delta_t, I_0, S_0, time, V_0)
    axes[i].plot(S, label='Susceptible', color='blue')
    axes[i].plot(I, label='Infected', color='red')
    axes[i].plot(R, label='Recovered', color='black')
    axes[i].plot(V, label='Vaccinated nothing', color='green')
    axes[i].set_xlabel("time")
    axes[i].set_ylabel("proportion of people")
    axes[i].set_title(f'R_0 = {beta}')
    axes[i].legend()

plt.tight_layout()
plt.savefig('Figures/ALM')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
#LM
time = 100
V_0 = v
for i, (beta, gamma) in enumerate(beta_gamma):
    S, I, R, V = forward_euler_LM(gamma, beta, delta_t, I_0, S_0, time, V_0, VE)
    axes[i].plot(S, label='Susceptible', color='blue')
    axes[i].plot(I, label='Infected', color='red')
    axes[i].plot(R, label='Recovered', color='black')
    axes[i].plot(V, label='Vaccinated', color='green')
    axes[i].set_xlabel("time")
    axes[i].set_ylabel("proportion of people")
    axes[i].set_title(f'R_0 = {beta}')
    axes[i].legend()

plt.tight_layout()
plt.savefig('Figures/LM')
plt.show()
```
