```python
import numpy as np
def forward_euler_grouping(D_s, D_p, C, D_w_inv, D_gam, time, s_0, i_0, delta_t):
    s = [np.array([[s_0],[s_0],[s_0],[s_0]])]
    i = [np.array([[i_0],[i_0],[i_0],[i_0]])]
    r = [np.array([[0],[0],[0],[0]])]
    
    for t in range(1,time):
        s_t = s[t-1] + delta_t * np.dot(-(D_s @ D_p @ C @ D_w_inv), i[t-1])
        s.append(s_t)
        
        i_t = i[t-1] + delta_t * (np.dot((D_s @ D_p @ C @ D_w_inv), i[t-1]) - np.dot(D_gam, i[t-1]))
        i.append(i_t)
        
        r_t = r[t-1] + delta_t * np.dot(D_gam, i[t-1])
        r.append(r_t)
    
    
    return s, i, r
```

```python
import matplotlib.pyplot as plt
c = 0.45
time = 100
s_0 = 0.99
i_0 = 0.01
delta_t = np.array([[0.1], [0.1], [0.1], [0.1]])
D_s = np.array([[0.2475, 0, 0, 0],
               [0, 0.2475, 0, 0],
               [0, 0, 0.2475, 0],
               [0, 0, 0, 0.2475]], dtype = 'float')
D_p = np.array([[1.0, 0, 0, 0],
               [0, 2.0, 0, 0],
               [0, 0, 3.0, 0],
               [0, 0, 0, 4.0]], dtype = 'float')
C = np.full((4, 4), c)
D_w_inv = np.array([[4.0, 0, 0, 0],
                   [0, 4.0, 0, 0],
                   [0, 0, 4.0, 0],
                   [0, 0, 0, 4.0]], dtype = 'float')
D_gam = np.array([[3.0, 0, 0, 0], 
                  [0, 3.0, 0, 0], 
                  [0, 0, 3.0, 0],
                  [0, 0, 0, 3.0]], dtype = 'float')
s, i, r = forward_euler_grouping(D_s, D_p, C, D_w_inv, D_gam, time, s_0, i_0, delta_t)

i_1, i_2, i_3, i_4 = [], [], [], []
for array_t in i:
    i_1.append(array_t[0][0])
    i_2.append(array_t[1][0])
    i_3.append(array_t[2][0])
    i_4.append(array_t[3][0])
```

```python
import matplotlib.colors as mcolors

# Plot 
base_color = 'red' 
base_hue = mcolors.rgb_to_hsv(mcolors.to_rgb(base_color))[0]
saturation_range = np.linspace(0.2, 1.0, 4)  

for i, data in enumerate([i_1, i_2, i_3, i_4]):
    hue = base_hue
    saturation = saturation_range[i]
    color = mcolors.hsv_to_rgb([hue, saturation, 1.0])
    plt.plot(data, label=f'Group {i+1}', color=color)

plt.xlabel('Time')
plt.ylabel('Infections')
plt.title('Population Infections')
plt.legend()
plt.savefig('Population_Infection')
plt.show()
```

```python
from scipy.stats import nbinom
from tabulate import tabulate 

def simulate_branching_process(R0, k, num_simulations=100000):
    mean = R0
    variance = mean + (mean**2)/k
    p = mean / variance
    n = mean**2 / (variance - mean)
    
    # Simulate num_simulations branching processes
    successes = 0
    for _ in range(num_simulations):
        # Continue branching until extinction or a large number of generations (in this case used 10)
        for _ in range(10):
            num_secondary_infections = nbinom.rvs(n=n, p=p)
            # If there are no secondary infections, the epidemic dies out
            if num_secondary_infections == 0:
                successes += 1
                break
    
    # Compute the probability of extinction
    extinction_probability = successes / num_simulations
    return extinction_probability
```

```python
# Values of R0 and k to consider
R0_values = [3]
k_values = [0.1, 0.5, 1.0, 5.0, 10.0]

table_data = []

# Loop through combinations of R0 and k
for R0 in R0_values:
    for k in k_values:
        # Simulate branching process and compute probability of extinction
        extinction_probability = simulate_branching_process(R0, k)
        table_data.append([R0, k, extinction_probability])

# Print table
print(tabulate(table_data, headers=["R0", "k", "Probability of Extinction"], floatfmt=".3f", tablefmt= "grid"))
```
